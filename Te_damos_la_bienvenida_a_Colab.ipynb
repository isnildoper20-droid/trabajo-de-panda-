```
# ============================================================
# ACTIVIDAD COMPLETA - LIBRERÍA PANDAS
# Autor: [Tu Nombre]
# ============================================================

# =========================
# 1. IMPORTAR LIBRERÍAS
# =========================

import pandas as pd
import numpy as np

print("Librerías importadas correctamente")


# ============================================================
# 2. ¿QUÉ ES PANDAS?
# ============================================================

print("\n¿Qué es Pandas?")
print("Pandas es una librería de Python utilizada para análisis y manipulación de datos.")


# ============================================================
# 3. CREAR UN DATAFRAME
# ============================================================

print("\n==============================")
print("CREACIÓN DE DATAFRAME")
print("==============================")

datos_estudiantes = {
    "Nombre": ["Ana", "Luis", "Carlos", "María"],
    "Edad": [20, 22, 21, 23],
    "Carrera": ["Ingeniería", "Medicina", "Derecho", "Arquitectura"]
}

df_estudiantes = pd.DataFrame(datos_estudiantes)

print(df_estudiantes)


# ============================================================
# 4. MOSTRAR INFORMACIÓN DEL DATAFRAME
# ============================================================

print("\n==============================")
print("INFORMACIÓN DEL DATAFRAME")
print("==============================")

print(df_estudiantes.info())


# ============================================================
# 5. OPERACIONES ENTRE COLUMNAS
# ============================================================

print("\n==============================")
print("OPERACIONES ENTRE COLUMNAS")
print("==============================")

productos = {
    "Producto": ["Mouse", "Teclado", "Monitor", "USB"],
    "Precio": [50000, 80000, 700000, 30000],
    "Cantidad": [2, 3, 1, 5]
}

df_productos = pd.DataFrame(productos)

# Crear nueva columna multiplicando precio por cantidad
df_productos["Total"] = df_productos["Precio"] * df_productos["Cantidad"]

print(df_productos)


# ============================================================
# 6. ESTADÍSTICAS BÁSICAS
# ============================================================

print("\n==============================")
print("ESTADÍSTICAS")
print("==============================")

print(df_productos.describe())


# ============================================================
# 7. ORDENAR DATOS
# ============================================================

print("\n==============================")
print("ORDENAR DATOS")
print("==============================")

ordenado = df_productos.sort_values(by="Total", ascending=False)

print(ordenado)


# ============================================================
# 8. FILTRAR DATOS
# ============================================================

print("\n==============================")
print("FILTRAR DATOS")
print("==============================")

filtro = df_productos[df_productos["Total"] > 100000]

print(filtro)


# ============================================================
# 9. CREAR ARCHIVO CSV
# ============================================================

print("\n==============================")
print("CREAR ARCHIVO CSV")
print("==============================")

df_productos.to_csv("productos.csv", index=False)

print("Archivo CSV creado correctamente")


# ============================================================
# 10. LEER ARCHIVO CSV
# ============================================================

print("\n==============================")
print("LEER ARCHIVO CSV")
print("==============================")

df_csv = pd.read_csv("productos.csv")

print(df_csv.head())


# ============================================================
# 11. DATOS SINTÉTICOS
# ============================================================

print("\n==============================")
print("DATOS SINTÉTICOS")
print("==============================")

datos_sinteticos = {
    "Estudiante": ["A", "B", "C", "D", "E"],
    "Nota": np.random.randint(1, 6, 5),
    "Asistencia": np.random.randint(70, 101, 5)
}

df_sinteticos = pd.DataFrame(datos_sinteticos)

print(df_sinteticos)


# ============================================================
# 12. AGREGAR NUEVA COLUMNA
# ============================================================

print("\n==============================")
print("NUEVA COLUMNA")
print("==============================")

df_sinteticos["Resultado"] = np.where(df_sinteticos["Nota"] >= 3,
                                      "Aprobó",
                                      "Reprobó")

print(df_sinteticos)


# ============================================================
# 13. DATOS NULOS
# ============================================================

print("\n==============================")
print("DATOS NULOS")
print("==============================")

datos_nulos = {
    "Nombre": ["Juan", "Pedro", None, "Laura"],
    "Edad": [20, None, 22, 19]
}

df_nulos = pd.DataFrame(datos_nulos)

print(df_nulos)

print("\nCantidad de datos nulos:")
print(df_nulos.isnull().sum())


# ============================================================
# 14. RELLENAR DATOS VACÍOS
# ============================================================

print("\n==============================")
print("RELLENAR DATOS VACÍOS")
print("==============================")

df_nulos["Edad"] = df_nulos["Edad"].fillna(df_nulos["Edad"].mean())

print(df_nulos)


# ============================================================
# 15. AGRUPAR DATOS
# ============================================================

print("\n==============================")
print("AGRUPAR DATOS")
print("==============================")

ventas = {
    "Vendedor": ["Ana", "Luis", "Ana", "Carlos", "Luis"],
    "Ventas": [300, 500, 200, 400, 350]
}

df_ventas = pd.DataFrame(ventas)

grupo = df_ventas.groupby("Vendedor")["Ventas"].sum()

print(grupo)


# ============================================================
# 16. CONCLUSIONES
# ============================================================

print("\n==============================")
print("CONCLUSIONES")
print("==============================")

print("""
1. Pandas facilita el análisis y manipulación de datos.
2. Permite crear tablas y trabajar con archivos CSV fácilmente.
3. Es muy importante en Inteligencia Artificial y Ciencia de Datos.
4. Ayuda a organizar y analizar grandes cantidades de información.
""")

print("\nActividad finalizada correctamente.")
```
